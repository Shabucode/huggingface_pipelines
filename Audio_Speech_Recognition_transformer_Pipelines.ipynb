#Transcription with custom input audio file

```python
from transformers import pipeline

# Initialize the ASR pipeline with a pre-trained model
asr_pipeline = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-large-960h")

# Path to the audio file (should be in WAV format, 16 kHz, mono)
audio_file_path = "/content/output_audio.wav"

# Transcribe the audio
transcription = asr_pipeline(audio_file_path)

# Output the transcription
print("Transcription:")
print(transcription["text"])
```
