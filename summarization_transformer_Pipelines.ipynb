```python
with pdfplumber.open('pdf1') as pdf:
  text = ''
  for page in pdf.pages:
    text+=page.extract_text()
```

```python
from transformers import pipeline

sum_pipeline = pipeline('summarization',model = 'facebook/bart-large-cnn', device=0 )
```

```python
import re
#cleaning text
text1 = re.sub(r'\s+',' ',text).strip()
```

```python
print(f"Length of story: {len(text)} characters")
```

```python
#chunck the story as BART model can handle onlu around 512 token~2000 chars
def chunk_text(text, max_characters=2000):
    """Split the text into smaller chunks of approximately max_characters."""
    chunks = []
    while len(text) > max_characters:
        # Find the last period within the limit
        split_at = text.rfind('.', 0, max_characters)
        if split_at == -1 or split_at < max_characters * 0.8:
            # If no period is found or period is too far, split at max_characters
            split_at = max_characters
        chunks.append(text[:split_at + 1].strip())
        text = text[split_at + 1:].strip()
    # Append any remaining text
    if text:
        chunks.append(text)
    return chunks
```

```python
chunks=chunk_text(text)
len(chunks)
```

```python
# Summarize each chunk using the BART model
summaries = []
for i, chunk in enumerate(chunks):
  print(f"summarizing chunk {i+1} of {len(chunks)}...")
  summary = sum_pipeline(chunk)[0]['summary_text']
  summaries.append(summary)
```

```python
# Combine all summaries
combined_summary = " ".join(summaries)

#summarize the combined summary for a shorter version
final_summary = sum_pipeline(combined_summary[:2000], max_length=150, min_length=50)[0]['summary_text']

print("Final Summary:", final_summary)
```
